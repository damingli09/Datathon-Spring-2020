```python
import pandas as pd
import csv
```

```python
df = pd.read_csv('weather.csv')
```

```python
df.head()
```

```python
df = df.loc[:,['DATE','DailyAverageWetBulbTemperature', 'DailyAverageWindSpeed','DailyPrecipitation','DailySnowfall']]
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['DATE'])
```

```python
df['start_weekday'] = df['date'].dt.weekday
```

```python
df.head()
```

```python
df = df[df['start_weekday'].isin([0,1,2,3,4])] # only weekdays
```

```python
df.to_csv('weather_clean.csv')
```
